```python
import mysql.connector as msql
from mysql.connector import Error
try:
    conn = msql.connect(host='localhost', database='stackoverflow', user='root', password='rootroot')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS stackoverflow_questions;')
        print('Creating table....')
        cursor.execute("CREATE TABLE stackoverflow_questions(id int,accepted_answer_id int,answer_count int,comment_count int,creation_date datetime,favorite_count int,owner_user_id int,related_posts varchar(255),score int,tags varchar(255),title varchar(255),view_count int, PRIMARY KEY (id))")
        print("Table is created")
        for i,row in SOFdata.iterrows():
            sql = "INSERT INTO SOF.stackoverflow_questions VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            conn.commit()
        print("Data is inserted")
except Error as e:
            print("Error while connecting to MySQL", e)
```
